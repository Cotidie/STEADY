# 1.2. Numpy Image Operation

### 1.2.1. Gray Scale
  
- Gray scale images are interpreted as a 2D array of values between 0 to 255.
- The value represents how dark a pixel is

### 1.2.2. RGB Image
   
 - RGB images contain 3 bytes per pixel (Intensities)
   - Each represents a gray scale for the color (Red, Green, Blue)



```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
pic = Image.open('images/00-puppy.jpg')
pic_arr = np.asarray(pic)
pic_arr.shape # 1300 x 1950 x 3 (RGB)
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()[:, :, 0]
pic_green = pic_arr.copy()[:, :, 1]
pic_blue = pic_arr.copy()[:, :, 2]

plt.imshow(pic_green, cmap='gray')
plt.show()
```

# 2. OpenCV
OpenCV is a Open Source library for real-time Computer Vision, developed by Intel. This contains many CV algorithms including object detection and tracking. 

OpenCV provides `imshow()` to open an image with a new window, but it interferes with jupytor kernel and makes it crash when exiting the window.
  - `cv2` provides fast and interactive window
    - suits for streaming/video/moving images  
  - Use `python` script instead

## 2.1. Image Basics

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
# 1. Opening an image
# By default, OpenCV opens images in BGR format (historical reason)
# To convert it to RGB, use cv2.cvtColor
pic = cv2.imread('images/00-puppy.jpg')
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
plt.show()
```

```python
img_gray = cv2.imread('images/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
# img_gray = cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# 2. Resizing an image
pic_small = cv2.resize(pic, (pic.shape[1]//10, pic.shape[0]//10))
plt.imshow(pic_small)
plt.show()
```

```python
# 3. Flipping an image
pic_flip = cv2.flip(pic, 1) # 0: vertical, 1: horizontal, -1: both
plt.imshow(pic_flip)
plt.show()
```

```python
# 4. Save an image
cv2.imwrite('images/00-puppy-gray.jpg', img_gray)
```

## 2.2. Drawing on Images
🔗 https://docs.opencv.org/4.x/dc/da5/tutorial_py_drawing_functions.html

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image_blank = np.zeros(shape=(512, 512, 3), dtype=np.uint16)
```

```python
cv2.rectangle(
    image_blank, 
    pt1=(100, 100), 
    pt2=(300, 300), 
    color=(255, 0, 0), 
    thickness=5
)
cv2.circle(
    image_blank, 
    center=(400, 400), 
    radius=10, 
    color=(0, 255, 0), 
    thickness=5
)
plt.imshow(image_blank)
plt.show()
```

```python
# Drawing text
# for more fonts: https://codeyarns.com/tech/2015-03-11-fonts-in-opencv.html#gsc.tab=0
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(
    image_blank, 
    text='Hello, OpenCV', 
    org=(10, 500), 
    fontFace=font, 
    fontScale=1, 
    color=(255, 255, 255), 
    thickness=2
)
plt.imshow(image_blank)
plt.show()
```
