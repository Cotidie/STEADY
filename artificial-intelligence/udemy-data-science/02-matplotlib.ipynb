# Matplotlib
| Link: http://matplotlib.org/  

Matplotlib allows you to create reproducible figures programmatically. It is an excellent 2D and 3D graphics library for generating scientific figures. `seaborn` is another plotting library built on top of `Matplotlib` that simplifies methods for various use-cases.
- Be able to plot functional relationship
- Be able to plot a relationship between raw data points

## Basics


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
y = 2*x

plt.title("y=2x")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xlim(0, 6)        # limit values from 0 to 6
plt.plot(x, y)
```

```python
# save to file
plt.savefig("y=2x.png")
```

## OOP

- **figure**: a canvas to hold multiple **Axes**, which is a single plots
- **Axes**: a subplot located inside of a **figure** , which contains Axis, Tick, Line2D, Text, Polygon, etc. and sets the coordinate system.

`subplots()` comes with pre-configured **Axes** for easily dividing sections across a **figure**

```python
# Call `plt.plot()` to display a figure

fig = plt.figure(
    dpi=200,          # resolution
    figsize=(12, 8),  # inches
)
axis_left = fig.add_axes([0.0, 0.0, 0.5, 1])
axis_right = fig.add_axes([0.5, 0.5, 0.5, 0.5])
axis_left.plot(x, y)
axis_right.plot(x, y)

# bounding box to 'tight' to resize the image to properly disply axis labels
fig.savefig('testing_bounding_box.png', bbox_inches='tight')
```

```python
# Use subplots to easily configure Axes
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
# plt.plot()
plt.tight_layout()    # to align subplots not to overlap
```

```python
# Adding legends
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x, label='y=x')
axes.plot(x, x**2, label='y=x^2')
axes.legend(loc=(1.1, 0.5))  # outside of plot, at the 50% of the height

plt.plot()
```

```python
# Adding styles (color, linewidth, linestyle, marker)
# refer to https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x**2, label='y=x^2', color='green', linewidth=2, linestyle='-', marker='o') 
axes.legend()
```
